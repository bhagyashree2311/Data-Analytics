# Q  Given marks of maths and stats for all students, First calculate the total marks and then find percentile

percentile = marks of a student / marks of the topper (highest marks)

Take array of marks for 5 students

```python
maths = np.array([87,95,85,88])
stats = np.array([93,90,84,80])
total_marks = maths+stats
percentile = (total_marks / total_marks.max())*100
```

```python
percentile
```

# Concatinate the arrays

```python
maths = np.array([87,95,85,88])
stats = np.array([93,90,84,80])
maths = maths.reshape(len(maths),1) # add a dummy dimension
stats = stats.reshape(len(stats),1) # add a dummy dimension
all_marks = np.concatenate((maths,stats),axis=1)
all_marks
```

```python
all_marks.sum(axis=1) # total marks of every student
```

```python
percentile = (all_marks.sum(axis=1) / all_marks.sum(axis=1).max() ) * 100
percentile
```

```python
import numpy as np

def neuron(I,W):
    X=sum(I*W)
    y = (1/(1+(np.exp(-X)))) # sigmoid activation function 1/ (1+e^(-x))
    return y
I=np.array([1,1,2,3,4])
W=np.array([2,2,2,2,2])
output=neuron(I,W)
print(output)
```
